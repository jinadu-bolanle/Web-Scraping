```python
lists = data.find_all("div", class_="info")

for item in lists:
        Name = item.h3.text
        print(Name)
        
```

```python
lists = data.find_all("div", class_="info")   # Extracting and storing the first page data 

header = ["Name", "Price", "Old_Price", "Discount", "Ratings"]

with open("iPhone1.csv", "w", encoding="UTF8", newline="",) as f:
    writer = csv.writer(f)
    writer.writerow(header)

    for item in lists:
        Name = item.find("h3", class_="name")
        Price = item.find("div", class_="prc")
        Old_Price = item.find("div", class_="old")
        Discount = item.find("div", class_="bdg _dsct _sm")
        Ratings = item.find("div", class_="rev")
        
        info = [Name, Price, Old_Price, Discount, Ratings]
        writer.writerow(info)
```

```python
lists = data.find_all("div", class_="info") # Extracting and storing the second page data

header = ["Name", "Price", "Old_Price", "Discount", "Ratings"]

with open("iPhone4.csv", "w", encoding="UTF8", newline="",) as f:
    writer = csv.writer(f)
    writer.writerow(header)

    for item in lists:
        Name = item.find("h3", class_="name")
        Price = item.find("div", class_="prc")
        Old_Price = item.find("div", class_="old")
        Discount = item.find("div", class_="bdg _dsct _sm")
        Ratings = item.find("div", class_="rev")
        
        info = [Name, Price, Old_Price, Discount, Ratings]
        writer.writerow(info)
```

```python
over_a_million = data[data['Price']>= 1000000]
half_a_million = data[(data['Price']>=500000) & (data['Price']<=999999)]
Higher_class = data[(data['Price']>= 200000) & (data['Price']<=499999)]
middle_class = data[(data['Price']>= 100000) & (data['Price']<=199999)]
lower_class = data[data['Price']<=99999]
```

```python
data1 = np.array([over_a_million['Price'].count(), half_a_million['Price'].count(), Higher_class['Price'].count(), middle_class['Price'].count(), lower_class['Price'].count()])
labels = ['>=1M', '>=0.5M & <=0.99M', '>=200K & <=499K', '>=100K & <=199K', '<100K']
plt.figure(figsize=(10,8))
plt.pie(data1, autopct = '%0.2F%%', labels = labels)
plt.title('Price distribution', fontsize = 16, fontweight='bold')
plt.show()
```
